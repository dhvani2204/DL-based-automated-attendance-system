```python
import cv2
import matplotlib.pyplot as plt

face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/VSEM Projects/DL Project/haarcascade_frontalface_default.xml')

image_path = '/content/drive/MyDrive/VSEM Projects/DL Project/classroom.jpeg'
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=4)

face_list = []

for (x, y, w, h) in faces:
    face_roi = gray_image[y:y+h, x:x+w]
    face_list.append(face_roi)
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

plt.figure(figsize=(12, 8))
for i, face in enumerate(face_list, 1):
    plt.subplot(1, len(face_list), i)
    plt.imshow(face, cmap='gray')
    plt.axis('off')
plt.show()
```
